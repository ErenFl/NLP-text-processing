```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import csv
```

```python
# Cargar el CSV en un DataFrame
ruta_archivo_entrada = "/content/normalized_data_corpus_lcd_293.csv"
ruta_archivo_salida = 'prueba2.csv'
```

```python
# Inicializar los vectorizadores
vectorizador_titulo = CountVectorizer(binary=True, token_pattern=r'(?u)\b\w\w+\b')
vectorizador_contenido = CountVectorizer(binary=True, token_pattern=r'(?u)\b\w\w+\b')

# Abrir el archivo CSV de entrada en modo lectura y el archivo de salida en modo escritura
with open(ruta_archivo_entrada, 'r', newline='', encoding='utf-8') as archivo_entrada, \
     open(ruta_archivo_salida, 'w', newline='', encoding='utf-8') as archivo_salida:

    # Crear un lector CSV para el archivo de entrada y un escritor CSV para el archivo de salida
    lector_csv = csv.reader(archivo_entrada)
    escritor_csv = csv.writer(archivo_salida)

    # Leer los encabezados y escribirlos en el archivo de salida
    encabezados = next(lector_csv)
    escritor_csv.writerow(encabezados)

    # Iterar sobre las filas restantes del archivo de entrada
    for fila in lector_csv:
        # Binarizar el título y el contenido
        titulo_binarizado = vectorizador_titulo.fit_transform([fila[2]]).toarray().tolist()[0]
        contenido_binarizado = vectorizador_contenido.fit_transform([fila[2]]).toarray().tolist()[0]

        # Escribir la fila binarizada en el archivo de salida
        fila_binarizada = [fila[0], titulo_binarizado, contenido_binarizado, fila[3], fila[4], fila[5]]
        escritor_csv.writerow(fila_binarizada)

print("Los datos normalizados se han guardado en 'prueba2.csv'.")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Texto a binarizar
texto = "Práctica común tiempo campaña electoral – ser cargo público aspirar – ser que candidato prometer ya ejercicio hueso soler incumplir , cuando menos parte medular interminable paquete “ compromiso ” “ oferta ” – notariada , ya ridículo , sangre medio – adquirido proselitismo . suceder aquí , allá acullá ."

# Inicializar el vectorizador binario
vectorizador = CountVectorizer(binary=True, token_pattern=r'(?u)\b\w\w+\b')

# Binarizar el texto
texto_binarizado = vectorizador.fit_transform([texto])

# Obtener el vocabulario
vocabulario = vectorizador.get_feature_names_out()

print("Texto binarizado:")
print(texto_binarizado.toarray())
print("\nVocabulario:")
print(vocabulario)
```
