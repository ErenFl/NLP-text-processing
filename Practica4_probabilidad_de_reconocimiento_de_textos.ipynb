```python
import csv
from collections import defaultdict

# Función para leer los corpus de bigramas desde un archivo CSV
def read_bigram_corpus(filename):
    bigram_corpus = defaultdict(int)
    with open(filename, 'r') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Saltar la primera fila (cabecera)
        for row in reader:
            terms = tuple(row[:-3])  # Ignorar las columnas de frecuencia y probabilidad condicional
            frequency = int(row[-3])
            bigram_corpus[terms] = frequency
    return bigram_corpus

# Función para calcular la probabilidad conjunta de un texto dado un corpus de bigramas con suavizado de Laplace
def calculate_joint_probability_with_laplace(text, bigram_corpus, smoothing=1):
    text = text.split()
    joint_probability = 1.0
    vocabulary_size = len(bigram_corpus)  # Tamaño del vocabulario (número de bigramas únicos)
    for i in range(len(text) - 1):
        bigram = tuple(text[i:i+2])
        bigram_count = bigram_corpus.get(bigram, 0)
        # Aplicar suavizado de Laplace
        probability = (bigram_count + smoothing) / (sum(bigram_corpus.values()) + smoothing * vocabulary_size)
        joint_probability *= probability
    return joint_probability

# Archivos CSV de corpus de bigramas
bigram_corpus_files = ['bigrams.csv', 'bigramsCruz.csv', 'bigrams_SSHM.csv', 'DUAG_Bigramas.csv']
#trigram_corpus_files = ['trigrams.csv', 'trigramsCruz.csv', 'trigrams_SSHM.csv', 'DUAG_Bigramas.csv']

# Leer corpus de bigramas
bigram_corpora = []
for filename in bigram_corpus_files:
    bigram_corpora.append(read_bigram_corpus(filename))

# Texto de ejemplo
input_text = "the quick brown fox jumps over the lazy dog"

# Calcular la probabilidad conjunta del texto dado los corpus de bigramas con suavizado de Laplace
for i, bigram_corpus in enumerate(bigram_corpora):
    joint_probability = calculate_joint_probability_with_laplace(input_text, bigram_corpus, smoothing=1)
    print(f"Probabilidad conjunta del texto con suavizado de Laplace (Bigramas {i+1}):", joint_probability)
```

```python
import csv
import tkinter as tk
from tkinter import filedialog
from collections import defaultdict

# Función para leer los corpus de bigramas desde un archivo CSV
def read_bigram_corpus(filename):
    bigram_corpus = defaultdict(int)
    with open(filename, 'r', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Saltar la primera fila (cabecera)
        for row in reader:
            terms = tuple(row[:-3])  # Ignorar las columnas de frecuencia y probabilidad condicional
            frequency = int(row[-3])
            bigram_corpus[terms] = frequency
    return bigram_corpus

# Función para leer los corpus de trigramas desde un archivo CSV
def read_trigram_corpus(filename):
    trigram_corpus = defaultdict(int)
    with open(filename, 'r', encoding='utf-8') as csvfile:
        reader = csv.reader(csvfile)
        next(reader)  # Saltar la primera fila (cabecera)
        for row in reader:
            terms = tuple(row[:3])  # Tomar los primeros tres elementos como términos del trigram
            frequency = int(row[3])  # La frecuencia del trigram está en la cuarta columna (índice 3)
            trigram_corpus[terms] = frequency
    return trigram_corpus


# Función para calcular la probabilidad conjunta de un texto dado un corpus con suavizado de Laplace
def calculate_joint_probability_with_laplace(text, corpus, smoothing=1):
    text = text.split()
    joint_probability = 1.0
    vocabulary_size = len(corpus)  # Tamaño del vocabulario (número de bigramas o trigramas únicos)
    for i in range(len(text) - len(next(iter(corpus)))):
        terms = tuple(text[i:i+len(next(iter(corpus)))])
        frequency = corpus.get(terms, 0)
        # Aplicar suavizado de Laplace
        probability = (frequency + smoothing) / (sum(corpus.values()) + smoothing * vocabulary_size)
        joint_probability *= probability
    return joint_probability

# Función para cargar archivos CSV y calcular las probabilidades
def load_and_calculate(bigram_corpus_files, trigram_corpus_files):
    # Limpiar resultados anteriores
    result_text.delete('1.0', tk.END)

    # Preguntar al usuario si desea utilizar bigramas o trigramas
    choice = corpus_choice.get()
    if choice == 'bigramas':
        corpus_files = bigram_corpus_files
    elif choice == 'trigramas':
        corpus_files = trigram_corpus_files
    else:
        result_text.insert(tk.END, "Opción no válida. Por favor, elige 'bigramas' o 'trigramas'.")
        return

    # Leer corpus seleccionado (bigramas o trigramas)
    corpora = []
    for filename in corpus_files:
        if choice == 'bigramas':
            corpus = read_bigram_corpus(filename)
        else:
            corpus = read_trigram_corpus(filename)
        corpora.append((filename, corpus))  # Guardar el nombre del archivo junto con el corpus

    # Texto de ejemplo
    input_text = input_text_entry.get()

    # Calcular la probabilidad conjunta del texto dado el corpus seleccionado con suavizado de Laplace
    probabilities = []
    for filename, corpus in corpora:
        joint_probability = calculate_joint_probability_with_laplace(input_text, corpus, smoothing=1)
        probabilities.append((filename, joint_probability))

    # Ordenar las probabilidades de mayor a menor
    sorted_probabilities = sorted(probabilities, key=lambda x: x[1], reverse=True)

    # Imprimir las probabilidades ordenadas
    for filename, prob_value in sorted_probabilities:
        result_text.insert(tk.END,f"Probabilidad conjunta ({filename}): {prob_value}\n")

# Función para abrir un cuadro de diálogo y permitir al usuario seleccionar archivos CSV
def browse_files():
    file_paths = filedialog.askopenfilenames(filetypes=[("CSV files", "*.csv")])
    if file_paths:
        for file_path in file_paths:
            file_list.insert(tk.END, file_path)

# Configuración de la interfaz gráfica
root = tk.Tk()
root.title("Calculadora de Probabilidades")

corpus_choice = tk.StringVar()
tk.Label(root, text="¿Deseas utilizar bigramas o trigramas?").pack()
tk.Radiobutton(root, text="Bigramas", variable=corpus_choice, value="bigramas").pack()
tk.Radiobutton(root, text="Trigramas", variable=corpus_choice, value="trigramas").pack()

input_text_entry = tk.Entry(root, width=50)
input_text_entry.pack()
input_text_entry.insert(0, "Texto de prueba")

file_list = tk.Listbox(root, selectmode=tk.MULTIPLE, width=50)
file_list.pack()

browse_button = tk.Button(root, text="Cargar modelos de lenguaje", command=browse_files)
browse_button.pack()

calculate_button = tk.Button(root, text="Calcular Probabilidades", command=lambda: load_and_calculate(file_list.get(0, tk.END), file_list.get(0, tk.END)))
calculate_button.pack()

result_text = tk.Text(root, height=20, width=120)
result_text.pack()

root.mainloop()
```
