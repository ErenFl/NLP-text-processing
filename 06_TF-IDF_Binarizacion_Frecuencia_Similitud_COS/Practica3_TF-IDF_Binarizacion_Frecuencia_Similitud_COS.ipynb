# **PROCESAMIENTO DEL CSV**



```python
import csv

# Cambiar path para local o Google Colab
path = "/content/normalized_data_corpus_lcd_293.csv"
#path = "normalized_data_corpus_lcd_293.csv"

ruta_archivo_entrada = path
ruta_archivo_salida = 'prueba3.csv'

# Lista para almacenar los títulos y contenidos
titulos = []
contenidos = []

# Abrir el archivo CSV en modo lectura
with open(ruta_archivo_entrada, 'r', newline='', encoding='utf-8') as archivo_csv:
    # Crear un lector CSV
    lector_csv = csv.reader(archivo_csv)

    # Leer cada fila del archivo CSV
    for fila in lector_csv:
        # Agregar el título y el contenido a las listas correspondientes
        titulos.append(fila[1])
        contenidos.append(fila[2])

# Imprimir todos los títulos en una lista
print("Titulos =", titulos)
archivo_salida_titulos = "titulos.txt"
with open(archivo_salida_titulos, 'w', encoding='utf-8') as archivo_txt:
    for titulo in titulos:
        archivo_txt.write(titulo + '\n')

print(f"Los títulos se han exportado correctamente en el archivo: {archivo_salida_titulos}")

# Imprimir todos los contenidos en una lista
print("Contenidos =", contenidos)
archivo_salida_contenidos = "contenidos.txt"
with open(archivo_salida_contenidos, 'w', encoding='utf-8') as archivo_txt:
    for contenido in contenidos:
        archivo_txt.write(contenido + '\n')

print(f"Los contenidos se han exportado correctamente en el archivo: {archivo_salida_contenidos}")

# Concatenar los títulos y contenidos
concatenados = [f"{titulos[i]} {contenidos[i]}" for i in range(len(titulos))]

print("Titulos + Contenidos =", concatenados)
archivo_salida_concatenados = "concatenados.txt"
with open(archivo_salida_concatenados, 'w', encoding='utf-8') as archivo_txt:
    for concatenado in concatenados:
        archivo_txt.write(concatenado + '\n')

print(f"Los títulos y contenidos concatenados se han exportado correctamente en el archivo: {archivo_salida_concatenados}")
```

# **STOPWORDS**

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from string import punctuation
from nltk.corpus import stopwords
import nltk

nltk.download('punkt')
nltk.download('stopwords')

# Definir stopwords en español y caracteres de puntuación
language_stopwords = set(stopwords.words('spanish'))
non_words = list(punctuation)

# Función para eliminar stopwords y caracteres de puntuación
def remove_stop_words(dirty_text):
    cleaned_text = ''
    for word in dirty_text.split():
        if word.lower() in language_stopwords or word in non_words:
            continue
        else:
            cleaned_text += word + ' '
    return cleaned_text

# Función para eliminar caracteres de puntuación
def remove_punctuation(dirty_string):
    for word in non_words:
        dirty_string = dirty_string.replace(word, '')
    return dirty_string

# Función para procesar un archivo de texto
def process_file(file_name):
    file_content = open(file_name, "r").read()
    # Convertir a minúsculas
    file_content = file_content.lower()
    # Eliminar caracteres de puntuación y stopwords
    file_content = remove_punctuation(file_content)
    file_content = remove_stop_words(file_content)
    return file_content
```

```python
# Procesar los tres archivos de texto
titulos_data = process_file("/content/titulos.txt")
content_data = process_file("/content/contenidos.txt")
concatenados_data = process_file("/content/concatenados.txt")
data_prueba = process_file("/content/Bellas Artes_recorte.txt")
```

**TF-IDF**

```python
# TF-IDF Vectorization con unigramas
tfidf_vectorizer_unigram = TfidfVectorizer(token_pattern=r'(?u)\b\w+\b|[^\w\s]',
                                           ngram_range=(1, 1))
X_tfidf_unigram = tfidf_vectorizer_unigram.fit_transform([titulos_data, data_prueba])
Y_tfidf_unigram = tfidf_vectorizer_unigram.fit_transform([content_data, data_prueba])
Z_tfidf_unigram = tfidf_vectorizer_unigram.fit_transform([concatenados_data, data_prueba])

# TF-IDF Vectorization con bigramas
tfidf_vectorizer_bigram = TfidfVectorizer(token_pattern=r'(?u)\b\w+\b|[^\w\s]',
                                          ngram_range=(2, 2))
X_tfidf_bigram = tfidf_vectorizer_bigram.fit_transform([titulos_data, data_prueba])
Y_tfidf_bigram = tfidf_vectorizer_bigram.fit_transform([content_data, data_prueba])
Z_tfidf_bigram = tfidf_vectorizer_bigram.fit_transform([concatenados_data, data_prueba])
```

# **BINARIZACION**

```python
# Binario Vectorization con unigramas
binary_vectorizer_unigram = CountVectorizer(binary=True, token_pattern=r'(?u)\b\w+\b|[^\w\s]', ngram_range=(1, 1))
X_bin_unigram = binary_vectorizer_unigram.fit_transform([titulos_data, data_prueba])
Y_bin_unigram = binary_vectorizer_unigram.fit_transform([content_data, data_prueba])
Z_bin_unigram = binary_vectorizer_unigram.fit_transform([concatenados_data, data_prueba])

# Binario Vectorization con bigramas
binary_vectorizer_bigram = CountVectorizer(binary=True, token_pattern=r'(?u)\b\w+\b|[^\w\s]', ngram_range=(2, 2))
X_bin_bigram = binary_vectorizer_bigram.fit_transform([titulos_data, data_prueba])
Y_bin_bigram = binary_vectorizer_bigram.fit_transform([content_data, data_prueba])
Z_bin_bigram = binary_vectorizer_bigram.fit_transform([concatenados_data, data_prueba])
```

# **FRECUENCIA**

```python
# Frecuencia Vectorization con unigramas
frequency_vectorizer_unigram = CountVectorizer(binary=False,token_pattern=r'(?u)\b\w+\b|[^\w\s]', ngram_range=(1, 1))
X_freq_unigram = frequency_vectorizer_unigram.fit_transform([titulos_data, data_prueba])
Y_freq_unigram = frequency_vectorizer_unigram.fit_transform([content_data, data_prueba])
Z_freq_unigram = frequency_vectorizer_unigram.fit_transform([concatenados_data, data_prueba])

# Frecuencia Vectorization con bigramas
frequency_vectorizer_bigram = CountVectorizer(binary=False, token_pattern=r'(?u)\b\w+\b|[^\w\s]', ngram_range=(2, 2))
X_freq_bigram = frequency_vectorizer_bigram.fit_transform([titulos_data, data_prueba])
Y_freq_bigram = frequency_vectorizer_bigram.fit_transform([content_data, data_prueba])
Z_freq_bigram = frequency_vectorizer_bigram.fit_transform([concatenados_data, data_prueba])
```

```python
# Calcular similitud de coseno

# TF-IDF unigramas
similarity_matrix_tf_unigram = cosine_similarity(X_tfidf_unigram, X_tfidf_unigram)
similarity_matrix_tf1_unigram = cosine_similarity(Y_tfidf_unigram, Y_tfidf_unigram)
similarity_matrix_tf2_unigram = cosine_similarity(Z_tfidf_unigram, Z_tfidf_unigram)

# Binario unigramas
similarity_matrix_b_unigram = cosine_similarity(X_bin_unigram, X_bin_unigram)
similarity_matrix_b1_unigram = cosine_similarity(Y_bin_unigram, Y_bin_unigram)
similarity_matrix_b2_unigram = cosine_similarity(Z_bin_unigram, Z_bin_unigram)

# Frecuencia unigramas
similarity_matrix_f_unigram = cosine_similarity(X_freq_unigram, X_freq_unigram)
similarity_matrix_f1_unigram = cosine_similarity(Y_freq_unigram, Y_freq_unigram)
similarity_matrix_f2_unigram = cosine_similarity(Z_freq_unigram, Z_freq_unigram)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer

# Definir los datos de titulos_data y data_prueba
titulos_data = ["texto de ejemplo 1", "texto de ejemplo 2"]
data_prueba = ["texto de prueba 1", "texto de prueba 2"]

# Inicializar el vectorizador TF-IDF unigrama
tfidf_vectorizer_unigram = TfidfVectorizer(token_pattern=r'(?u)\b\w+\b|[^\w\s]', ngram_range=(1, 1))

# Ajustar el vectorizador a ambos conjuntos de datos y transformarlos en matrices TF-IDF
tfidf_combined = tfidf_vectorizer_unigram.fit_transform(titulos_data + data_prueba)

# Obtener las dimensiones de los conjuntos de datos originales
n_titulos = len(titulos_data)
n_prueba = len(data_prueba)

# Separar las matrices TF-IDF
a1 = tfidf_combined[:n_titulos]
a2 = tfidf_combined[n_titulos:]

# Calcular la similitud del coseno entre los vectores TF-IDF
from sklearn.metrics.pairwise import cosine_similarity
similarity_matrix = cosine_similarity(a1, a2)

# Imprimir la matriz de similitud
print(similarity_matrix)
```
